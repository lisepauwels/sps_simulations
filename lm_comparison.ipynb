```python
def get_lossmap_paths(path, file_name_base):

    n_jobs = int(subprocess.check_output(f'find {path}/job* -maxdepth 0 -type d | wc -l', shell=True))
    n_jobs_succeeded = int(subprocess.check_output(f'find {path}/job* -maxdepth 0 -type d -not -empty | wc -l', shell=True))
    
    job_files_list = []
    for i in range(n_jobs):
        current_file_path = (subprocess.check_output(f'echo {path}/job_{i}/{file_name_base}.json', shell=True)).decode('ascii').strip()
        #current_file_path = (subprocess.check_output(f'echo {path}/job_{i}/{file_name_base}_{plane}.json', shell=True)).decode('ascii').strip()
        
        if os.path.exists(current_file_path):
            job_files_list.append(current_file_path)
        else:
            print(f'Job {i} failed')
    
    print(f'{n_jobs_succeeded} out of a total of {n_jobs} succeeded')
    if n_jobs_succeeded != len(job_files_list):
        print('ERROR: not all succeeded jobs paths were retrieved')
        
    return job_files_list
```

```python
combined_data = {}

for sim_type in ['ripple_errors', 'ripple', 'errors']:
    for shell in ['xshell', 'yshell', 'lshell', 'xlshell', 'ylshell', 'xyshell', 'xylshell']:
        print(f'{sim_type}_{shell}')
        
        loc_dic = {
            'collimator': {'s': [], 'name': [], 'length': [], 'n': []},
            'aperture': {'s': [], 'name': [], 'n': []},
            'machine_length': 6911.5038,
            'interpolation': 0.1,
            'reversed': False
            }

        for i in range(len(lms_paths[sim_type][shell])):
            with open(lms_paths[sim_type][shell][i], "r") as f:
                data=json.load(f)

            #Part collimator
            if len(loc_dic['collimator']['s']) == 0:
                loc_dic['collimator']['s'] = data['collimator']['s']
                loc_dic['collimator']['name'] = data['collimator']['name']
                loc_dic['collimator']['length'] = data['collimator']['length']
                loc_dic['collimator']['n'] = data['collimator']['n']

            else:
                if len(data['collimator']['n']) > 0:
                    loc_dic['collimator']['n'][0] += data['collimator']['n'][0]

            #Part aperture
            if len(loc_dic['aperture']['s']) == 0:
                loc_dic['aperture']['s'] = data['aperture']['s']
                loc_dic['aperture']['name'] = data['aperture']['name']
                loc_dic['aperture']['n'] = data['aperture']['n']

            else:
                for j in range(len(data['aperture']['name'])):
                    name = data['aperture']['name'][j]
                    if name in loc_dic['aperture']['name']:
                        idx = loc_dic['aperture']['name'].index(name)
                        loc_dic['aperture']['n'][idx] += data['aperture']['n'][j]
                    else:
                        loc_dic['aperture']['name'].append(name)
                        loc_dic['aperture']['n'].append(data['aperture']['n'][j])
                        loc_dic['aperture']['s'].append(data['aperture']['s'][j])
        
        combined_data[f'{sim_type}_{shell}'] = loc_dic
```

```python
for sim_type in combined_data:
    combined_data[sim_type]['aperture']['s'] = np.array(combined_data[sim_type]['aperture']['s'])
    combined_data[sim_type]['aperture']['n'] = np.array(combined_data[sim_type]['aperture']['n'])
```

```python
xi = {'xi1' : {'position' : [775,850]},
      'xi2' : {'position' : [5220, 5500]},
      'xi3' : {'position' : [5650, 5800]},
      'xi4' : {'position' : [6475, 6710]}
    }

for sim_type in combined_data:
    for xi_key in xi:
        lower = np.where(combined_data[sim_type]['aperture']['s'] > xi[xi_key]['position'][0])
        upper = np.where(combined_data[sim_type]['aperture']['s'] < xi[xi_key]['position'][1])
        idxs = np.intersect1d(lower, upper)
        
        if len(idxs) > 0:
            if len(combined_data[sim_type]['aperture']['n']) > 0 and len(combined_data[sim_type]['collimator']['n']) > 0:
                xi[xi_key][sim_type] = np.sum(combined_data[sim_type]['aperture']['n'][idxs])/(np.sum(combined_data[sim_type]['aperture']['n']) + combined_data[sim_type]['collimator']['n'][0])
            else:
                if len(combined_data[sim_type]['collimator']['n']) > 0 and len(combined_data[sim_type]['aperture']['n']) == 0:
                    xi[xi_key][sim_type] = np.sum(combined_data[sim_type]['aperture']['n'][idxs])/combined_data[sim_type]['collimator']['n'][0]
                elif len(combined_data[sim_type]['collimator']['n']) == 0 and len(combined_data[sim_type]['aperture']['n']) > 0:
                    xi[xi_key][sim_type] = np.sum(combined_data[sim_type]['aperture']['n'][idxs])/np.sum(combined_data[sim_type]['aperture']['n'])
        
        else:
            xi[xi_key][sim_type] = 0
```
