# Intensity loss

```python
def combine_intensity_loss(paths):
    combined_data = {}

    for p in paths:
        with open(p, 'r') as fid:
            data = json.load(fid)
        
        for key in sorted(list(data.keys())):
            if key != 'sweep_per_turn':
                if key not in combined_data:
                    combined_data[key] = data[key]
                else:
                    combined_data[key] += data[key]
            
            else:
                if key in combined_data:
                    assert combined_data[key] == data[key], 'Sweep per turn is not the same, is it the same simulation type?'
                else:
                    combined_data[key] = data[key]
    return combined_data
```

# loss locations

```python
def combine_loss_locations(paths):
    combined_data = {}

    for p in paths:
        with open(p, 'r') as fid:
            data = json.load(fid)
        
        for key in data:
            if key not in combined_data:
                combined_data[key] = data[key]
            else:
                combined_data[key] += data[key]

    return combined_data
```

```python
studies_dir = Path('/eos/user/l/lpauwels/Simulations/OffMomBumpScans/studies_result/')
for quad in subdirs:
    for plane in subdirs[quad]:
        for bump in subdirs[quad][plane]:
            paths = [Path(p/'loss_locations.json') for p in subdirs[quad][plane][bump]]
            savedir = Path(studies_dir/quad/plane/bump)
            savedir.mkdir(parents=True, exist_ok=True)
            combined_data = combine_loss_locations(paths)
            json.dump(combined_data, open(Path(savedir/'loss_locations.json'), 'w'), indent=4)
```
