```python
import numpy as np
import awkward as ak
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colormaps
from pathlib import Path

import xtrack as xt
import xcoll as xc
```

```python
line = xt.load('../injection_lines/sps_with_aperture_inj_q20_beam_sagitta4.json')
```

```python
twisses = {-0.01: line.twiss4d(delta0=-0.01), 0.0: line.twiss4d(delta0=0.0), 0.01: line.twiss4d(delta0=0.01)}
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
for delta0, twiss in twisses.items():
    ax.plot(twiss.s, twiss.dx, label=f'delta0={delta0}')
ax.set_xlabel('s [m]')
ax.set_ylabel('dx [m]')
ax.legend(loc='lower right')
ax.grid()
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
ax.plot(twiss.s, (twisses[0.01].dx-twisses[0.0].dx)/twisses[0.0].dx, label=f'delta0={0.01}')
ax.plot(twiss.s, (twisses[-0.01].dx-twisses[0.0].dx)/twisses[0.0].dx, label=f'delta0={-0.01}')
ax.set_xlabel('s [m]')
ax.set_ylabel('dx [m]')
ax.legend(loc='lower right')
ax.grid()
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
ax.plot(twiss.s, (twisses[0.01].betx-twisses[0.0].betx)/twisses[0.0].betx, label=f'delta0={0.01}')
ax.plot(twiss.s, (twisses[-0.01].betx-twisses[0.0].betx)/twisses[0.0].betx, label=f'delta0={-0.01}')
ax.set_xlabel('s [m]')
ax.set_ylabel('beta_x [m]')
ax.legend(loc='lower right')
ax.grid()
```
