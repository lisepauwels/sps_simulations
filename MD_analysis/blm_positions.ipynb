```python
line = xt.Line.from_json('../injection_lines/sps_with_aperture_inj_q20_beam_sagitta.json')
```

```python
element_names = []
for name in line.element_names[2:-2]:
    if 'drift' not in name:
        element_names.append(name)
```

```python
tt = line.get_table()
```

```python
elems_types = []
for elem in line:
    elems_types.append(elem.__class__.__name__)
np.unique(np.array(elems_types))
```

```python
elems, names = line.get_elements_of_type([xt.Cavity, xt.Multipole, xt.Octupole, xt.Quadrupole, 
                                          xt.RFMultipole, xt.Sextupole, xt.ThickSliceOctupole, 
                                          xt.ThickSliceQuadrupole, xt.ThickSliceRBend, xt.ThickSliceSextupole])
```

```python
line_numbers_restr = []
for name in names:
    line_numbers_restr.append(int(name.split('.')[1]))

blm_numbers = []
for name in blm_names:
    number = int(name.split('.')[2])
    blm_numbers.append(number)
```

```python
sorted_numbers, sorted_names = zip(*sorted(zip(line_numbers_restr, names)))
```

```python
import bisect
```

```python
blm_to_downstream = {
    blm_name: names[idx] if idx < len(names) else None
    for blm_name, blm_number in zip(blm_names, blm_numbers)
    for idx in [bisect.bisect_right(sorted_numbers, blm_number)]
}
```

```python
ntd = []
for name in blm_to_downstream:
    if blm_to_downstream[name] is None:
        ntd.append(name)
for name in ntd:
    del blm_to_downstream[name]
```

```python
blm_positions = {name : line.get_s_position(blm_to_downstream[name]) for name in blm_to_downstream}
```
